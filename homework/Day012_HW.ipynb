### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup
from collections import Counter

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
all_data = {'date':[],'type':[],'title':[]}

for d in soup.find(class_="part_list_2").find_all('h3'):
    all_data['date'].append(d.find(class_="date").text)
    all_data['type'].append(d.find_all('em')[-1].text)
    all_data['title'].append(d.find_all('a')[-1].text.replace(u'\u3000', ''))

print('類別種類:',Counter(all_data['type']).keys())
print(all_data)
```

### ③ 哪一個類別的文章最多

```python
data = Counter(all_data['type']).most_common(1)
mosttype = data[0][0]
value = data[0][1]
print(mosttype,'類的文章最多, 共',value,'篇')

```
